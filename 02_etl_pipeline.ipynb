```python
import pandas as pd
import requests as rq #you need to access the web using this lib
import sqlite3 as sql


# Extracting:

url = 'https://en.wikipedia.org/wiki/List_of_largest_banks'
response = rq.get(url, headers={'User-Agent': 'Mozilla/5.0'})
table = pd.read_html(response.text)
df = table[0] #it has 3 tables(len(df) and you'll know) so we only need one

# Cleaning:
df = df.rename(columns={"Total assets (December 2025) (US$ billion)": "TDD"})

# Loading:

conn = sql.connect("ETL.db")
df.to_sql("Largest_Banks", conn, if_exists='replace',index=False)
conn.close()
```
